6. Data cleaning

- Handling missing values

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
for i in df_reviews.columns :
    if df_reviews[i].dtype in ['int64','float64']:
        df_reviews[i].fillna(value=df_reviews[i].median(),inplace=True)
    else :
        df_reviews[i].fillna(value=df_reviews[i].mode()[0],inplace=True)
```

- Handling outliers

```python
q1={}
q3={}
outliers={}
for i in df_reviews.columns :
  if df_reviews[i].dtype in ['float64','int64']:
    q1[i]=df_reviews[i].quantile(0.25)
    q3[i]=df_reviews[i].quantile(0.75)
    iqr=q3[i]-q1[i]
    outliers[i]=df_reviews[(df_reviews[i]>=q3[i]+iqr*1.5)|(df_reviews[i]<=q1[i]-1.5*iqr)]
    print(f'The percentage of outliers in {i} columns is {len(outliers[i])*100/len(df_reviews):.2f}%')
```

> All outliers percentages are less than 1 % so we are going to keep them .

- Encoding categorical values

```python
from sklearn.preprocessing import LabelEncoder
encoders = {}
for col in filter(lambda x: df_reviews[x].dtype == 'object', df_reviews.columns):
    le = LabelEncoder()
    df_reviews[col] = le.fit_transform(df_reviews[col])
    encoders[col] = le
```

- Predicting overall rating  

Splitting the data between test and train

```python
from sklearn.model_selection import train_test_split
X=df_reviews.drop(columns=['food_beverages','header','author','recommended','rating_range','rating',"seat_comfort",'cabin_staff_service','food_beverages','ground_service','value_for_money','entertainment','date','content'])
y=df_reviews['rating_range']
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.4)
```

- Scaling data

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
X_train_scaled = s.fit_transform(X_train)
X_test_scaled = s.transform(X_test)
```

Model selection + Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score
import pandas as pd
import numpy as np

models = {
    'LogisticRegression': LogisticRegression(C=3, penalty='l2', solver='lbfgs', max_iter=2000),
    'SVC': SVC(C=3,kernel='rbf',gamma='scale'),
    'XGBoost': XGBClassifier(n_estimators=300,learning_rate=0.05,max_depth=6,subsample=0.8,colsample_bytree=0.8,reg_alpha=0.1,reg_lambda=1,min_child_weight=3,objective='multi:softprob',num_class=len(np.unique(y_train)),random_state=42)
}

results = []
trained_models = {}

for name, model in models.items():
    print(f'Training {name}...')
    model.fit(X_train, y_train)
    trained_models[name] = model

    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')

    results.append({
        'Model': name,
        'Accuracy': acc,
        'F1 Score': f1
    })

results_df = pd.DataFrame(results)
print(results_df.sort_values(by='Accuracy', ascending=False))
```

> All of the models that we have tried didn't lead to a good performance so we are going to try a new method 'Models Stacking ' .

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
X_train,y_train=smote.fit_resample(X_train,y_train)
```

```python
from sklearn.ensemble import StackingClassifier

stack10 = StackingClassifier(estimators=[
    ('lr', LogisticRegression(C=0.3, solver='lbfgs', max_iter=2000)),
    ('svc', SVC(C=1, kernel='rbf', gamma=0.2, probability=True)),
    ('xgb', XGBClassifier(
        n_estimators=600, learning_rate=0.07, max_depth=9,
        subsample=0.8, colsample_bytree=0.9,
        reg_alpha=0.2, reg_lambda=1.3, min_child_weight=2,
        objective='multi:softprob', num_class=len(np.unique(y_train)), random_state=42))
], final_estimator=LogisticRegression(C=0.5))
stack10.fit(X_train,y_train)
y_pred=stack10.predict(X_test)
```

```python
stack1 = StackingClassifier(estimators=[
    ('lr', LogisticRegression(class_weight='balanced', C=3, penalty='l2', solver='lbfgs', max_iter=2000)),
    ('svc', SVC(class_weight='balanced', C=5, kernel='rbf', gamma='scale', probability=True)),
    ('xgb', XGBClassifier(
        n_estimators=500, learning_rate=0.05, max_depth=6, subsample=0.8,
        colsample_bytree=0.8, reg_alpha=0.1, reg_lambda=1,
        min_child_weight=3, objective='multi:softprob',
        num_class=len(np.unique(y_train)), random_state=42))
], final_estimator=LogisticRegression(max_iter=1000))
stack2 = StackingClassifier(estimators=[
    ('lr', LogisticRegression(C=1.5, penalty='l2', solver='lbfgs', max_iter=2000)),
    ('svc', SVC(C=3, kernel='rbf', gamma=0.5, probability=True)),
    ('xgb', XGBClassifier(
        n_estimators=700, learning_rate=0.07, max_depth=8,
        subsample=0.9, colsample_bytree=0.8,
        reg_alpha=0.05, reg_lambda=1, min_child_weight=2,
        objective='multi:softprob', num_class=len(np.unique(y_train)), random_state=42))
], final_estimator=LogisticRegression(C=2))
stack4 = StackingClassifier(estimators=[
    ('lr', LogisticRegression(C=2, solver='lbfgs', max_iter=2000)),
    ('svc', SVC(C=2, kernel='sigmoid', gamma='scale', probability=True)),
    ('xgb', XGBClassifier(
        n_estimators=300, learning_rate=0.03, max_depth=4,
        subsample=0.75, colsample_bytree=0.7,
        reg_alpha=0.2, reg_lambda=0.8, min_child_weight=5,
        objective='multi:softprob', num_class=len(np.unique(y_train)), random_state=42))
], final_estimator=LogisticRegression())
stack5 = StackingClassifier(estimators=[
    ('lr', LogisticRegression(C=0.5, penalty='l1', solver='saga', max_iter=2000)),
    ('svc', SVC(C=1, kernel='linear', probability=True)),
    ('xgb', XGBClassifier(
        n_estimators=600, learning_rate=0.05, max_depth=6,
        subsample=0.8, colsample_bytree=0.8,
        reg_alpha=0.3, reg_lambda=1.5, min_child_weight=3,
        objective='multi:softprob', num_class=len(np.unique(y_train)), random_state=42))
], final_estimator=LogisticRegression(C=0.8))
```

```python
stacks = [stack1, stack2, stack4, stack5]
results = []

for i, model in enumerate(stacks, 1):
    print(f"Training Stack {i}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    results.append({'Model': f'Stack{i}', 'Accuracy': acc, 'F1': f1})

results_df = pd.DataFrame(results)
print(results_df.sort_values(by='Accuracy', ascending=False))
```
